```python
import os
import pandas as pd
import geopandas as gpd
import glob
from datetime import datetime

base_dir = "/Users/leo/Desktop/CNGF5020/MINIPRO/Mini Group Project I Data(1)"
fire_dir = os.path.join(base_dir, "Satellite Fire Data")
county_shp_path = os.path.join(base_dir, "CHN_County.shp")

fire_csv_paths = glob.glob(os.path.join(fire_dir, "modis_*.csv"))
fire_csv_paths.sort()
print(f"找到{len(fire_csv_paths)}个年份的火点数据文件：")
for path in fire_csv_paths:
    print(f"- {os.path.basename(path)}")


def get_heilongjiang_boundary(shp_path):
    county_gdf = gpd.read_file(shp_path)
    province_field = "NAME_1"
    print(f"\n行政边界数据的省份字段名：{province_field}")
    print(f"数据中包含的省份示例：{county_gdf[province_field].unique()[:5]}")
    
    hlj_county_gdf = county_gdf[county_gdf[province_field] == "Heilongjiang"]
    if len(hlj_county_gdf) == 0:
        hlj_county_gdf = county_gdf[county_gdf[province_field] == "HEILONGJIANG"]
        if len(hlj_county_gdf) == 0:
            raise ValueError("未找到黑龙江省数据！请检查省份字段值大小写")
    
    print(f"\n黑龙江省（Heilongjiang）包含{len(hlj_county_gdf)}个县级行政单元")
    return hlj_county_gdf

hlj_boundary_gdf = get_heilongjiang_boundary(county_shp_path)


def preprocess_fire_data(csv_path):
    file_name = os.path.basename(csv_path)
    year = file_name.split("_")[1]
    
    try:
        core_fields = ['latitude', 'longitude', 'acq_date', 'acq_time', 'frp', 
                       'confidence', 'daynight', 'type']
        df = pd.read_csv(csv_path, usecols=core_fields)

        df['acq_time_str'] = df['acq_time'].astype(str).str.zfill(4)
        df['acq_time_formatted'] = df['acq_time_str'].str[:2] + ":" + df['acq_time_str'].str[2:]

        df = df[
            (df['latitude'] >= 43) & (df['latitude'] <= 53) &
            (df['longitude'] >= 121) & (df['longitude'] <= 135)
        ]
        df = df[df['frp'] > 0]
        df = df.dropna(subset=['latitude', 'longitude', 'acq_date', 'acq_time_formatted', 'frp'])
        
        if len(df) == 0:
            print(f"⚠️ {year}年数据清洗后无有效记录，跳过")
            return None
        

        df['datetime'] = pd.to_datetime(
            df['acq_date'] + " " + df['acq_time_formatted'],
            format="%Y-%m-%d %H:%M",  
            errors="coerce"
        )
        df = df.dropna(subset=['datetime'])  
        

        df['year'] = year
        df['month'] = df['datetime'].dt.month  
        df['week'] = df['datetime'].dt.isocalendar().week  
        df['hour'] = df['datetime'].dt.hour  
        
        df = df.drop(columns=['acq_time', 'acq_time_str', 'acq_time_formatted'])
        
        print(f" {year}年数据处理完成：原始{pd.read_csv(csv_path).shape[0]}条 → 有效{len(df)}条")
        return df
    
    except Exception as e:
        print(f"❌ {year}年数据处理失败：{str(e)}")
        return None


fire_df_list = []
for csv_path in fire_csv_paths:
    processed_df = preprocess_fire_data(csv_path)
    if processed_df is not None:
        fire_df_list.append(processed_df)

if len(fire_df_list) == 0:
    raise ValueError("所有年份数据处理后均无有效记录！请检查CSV文件是否损坏")

# 合并所有年份数据
fire_all_df = pd.concat(fire_df_list, ignore_index=True)
print(f"\n📊 所有年份数据合并完成：共{len(fire_all_df)}条有效火点记录（2010-2019）")

# 转换为GeoDataFrame（WGS84坐标系）
fire_gdf = gpd.GeoDataFrame(
    fire_all_df,
    geometry=gpd.points_from_xy(fire_all_df['longitude'], fire_all_df['latitude']),
    crs="EPSG:4326"
)

# 裁剪：只保留黑龙江省内的火点
fire_hlj_gdf = gpd.clip(fire_gdf, hlj_boundary_gdf)
print(f"🌍 黑龙江省内火点数量：{len(fire_hlj_gdf)}条")


print("\n黑龙江省火点数据前5行（核心字段）：")
display(fire_hlj_gdf[['latitude', 'longitude', 'datetime', 'frp', 'year', 'month', 'geometry']].head())

yearly_count = fire_hlj_gdf.groupby('year').size().reset_index(name='fire_count').sort_values('year')
print("\n黑龙江省各年份火点数量统计：")
display(yearly_count)

monthly_count = fire_hlj_gdf.groupby('month').size().reset_index(name='fire_count').sort_values('month')
print("\n黑龙江省各月份火点数量统计（看季节分布）：")
display(monthly_count)
```

```python
import os
import pandas as pd
import geopandas as gpd
import rasterio
from rasterio.transform import rowcol
import matplotlib.pyplot as plt
import numpy as np
```

```python
def match_fire_with_crop(fire_gdf, crop_tif_paths, crop_type="maize"):
    """
    将火点数据与作物物候数据匹配，计算时间差
    fire_gdf: 火点GeoDataFrame（含datetime和year字段）
    crop_tif_paths: 作物物候TIFF文件路径列表
    crop_type: 作物类型（"maize"或"wheat"）
    return: 匹配后的火点GeoDataFrame（含作物成熟日和时间差）
    """
    matched_dfs = []
    # 遍历每一年的作物数据
    for tif_path in crop_tif_paths:
        # 提取年份（从文件名中解析，如"Heilongjiang_Maize_MA_2010.tif" → 2010）
        year = int(os.path.basename(tif_path).split("_")[-1].split(".")[0])
        print(f"处理{year}年{crop_type}数据...")
        
        # 筛选该年份的火点
        fire_year = fire_gdf[fire_gdf["year"].astype(int) == year].copy()
        if len(fire_year) == 0:
            print(f"⚠️ {year}年无火点数据，跳过")
            continue
        
        # 读取作物成熟日TIFF
        with rasterio.open(tif_path) as src:
            # 获取栅格元数据（坐标系、转换参数）
            transform = src.transform
            nodata = src.nodata
            # 计算火点在栅格中的行列号
            rows, cols = rowcol(
                transform, 
                fire_year.geometry.x,  # 火点经度
                fire_year.geometry.y   # 火点纬度
            )
            
            # 提取成熟日DOY（排除nodata区域，即非作物种植区）
            maturity_doy = []
            for r, c in zip(rows, cols):
                # 检查行列号是否在栅格范围内
                if 0 <= r < src.height and 0 <= c < src.width:
                    val = src.read(1)[r, c]
                    maturity_doy.append(val if val != nodata else None)
                else:
                    maturity_doy.append(None)
            
            # 添加成熟日字段
            fire_year[f"{crop_type}_maturity_doy"] = maturity_doy
            # 过滤非作物种植区的火点（成熟日为None的）
            fire_crop = fire_year.dropna(subset=[f"{crop_type}_maturity_doy"])
            
            # 计算火点DOY（年积日）
            fire_crop["fire_doy"] = fire_crop["datetime"].dt.dayofyear
            # 计算时间差（火点DOY - 成熟日DOY）
            fire_crop[f"days_after_{crop_type}"] = fire_crop["fire_doy"] - fire_crop[f"{crop_type}_maturity_doy"]
            
            matched_dfs.append(fire_crop)
    
    # 合并所有年份数据
    if not matched_dfs:
        return None
    return pd.concat(matched_dfs, ignore_index=True)
```

```python
# 匹配玉米数据
fire_maize_gdf = match_fire_with_crop(
    fire_gdf=fire_hlj_gdf,
    crop_tif_paths=maize_tif_paths,
    crop_type="maize"
)

# 匹配小麦数据
fire_wheat_gdf = match_fire_with_crop(
    fire_gdf=fire_hlj_gdf,
    crop_tif_paths=wheat_tif_paths,
    crop_type="wheat"
)

# 检查结果
if fire_maize_gdf is not None:
    print(f"\n玉米匹配结果：{len(fire_maize_gdf)}条火点位于玉米种植区")
if fire_wheat_gdf is not None:
    print(f"小麦匹配结果：{len(fire_wheat_gdf)}条火点位于小麦种植区")
```

```python
def analyze_time_diff(matched_gdf, crop_type="maize"):
    """分析火点与作物成熟日的时间差分布"""
    diff_col = f"days_after_{crop_type}"
    # 过滤时间差在合理范围的样本（例如：成熟前30天到成熟后60天）
    valid_diff = matched_gdf[
        (matched_gdf[diff_col] >= -30) & 
        (matched_gdf[diff_col] <= 60)
    ]
    
    print(f"\n{crop_type}时间差统计（单位：天）：")
    print(f"均值：{valid_diff[diff_col].mean():.1f}")
    print(f"中位数：{valid_diff[diff_col].median()}")
    print(f"成熟后1-30天内的火点占比：{((valid_diff[diff_col] >=1) & (valid_diff[diff_col] <=30)).mean():.2%}")
    return valid_diff

# 分析玉米
if fire_maize_gdf is not None:
    maize_valid = analyze_time_diff(fire_maize_gdf, "maize")

# 分析小麦
if fire_wheat_gdf is not None:
    wheat_valid = analyze_time_diff(fire_wheat_gdf, "wheat")
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

font_path = "/System/Library/AssetsV2/com_apple_MobileAsset_Font7/62032b9b64a0e3a9121c50aeb2ed794e3e2c201f.asset/AssetData/Hei.ttf"

# 验证字体路径是否存在
if os.path.exists(font_path):
    font_prop = fm.FontProperties(fname=font_path)
    plt.rcParams["font.family"] = font_prop.get_name()
else:
    # 若指定字体不存在，尝试系统默认中文字体
    plt.rcParams["font.family"] = ["Arial Unicode MS", "Heiti TC", "SimHei"]

plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题


# --------------------------
# 2. 优化绘图函数（增加数据验证）
# --------------------------
def plot_time_diff(valid_diff, crop_type="maize"):
    """绘制时间差直方图（火点与作物成熟日）"""
    # 数据有效性验证
    if valid_diff is None or valid_diff.empty:
        print(f"⚠️ 无有效{crop_type}数据，无法绘图")
        return
    
    diff_col = f"days_after_{crop_type}"
    # 检查必要列是否存在
    if diff_col not in valid_diff.columns:
        print(f"⚠️ 数据中缺少必要列：{diff_col}，无法绘图")
        return
    
    # 绘图
    plt.figure(figsize=(10, 6))
    # 绘制直方图（ bins=30 分组）
    n, bins, patches = plt.hist(
        valid_diff[diff_col], 
        bins=30, 
        color="#FF6B6B", 
        edgecolor="white",
        alpha=0.7
    )
    # 标记关键区间（成熟后1-30天）
    plt.axvspan(1, 30, color="#4ECDC4", alpha=0.3, label="成熟后1-30天（高风险期）")
    
    # 中文标签（根据作物类型显示对应名称）
    crop_name = "玉米" if crop_type == "maize" else "小麦"
    plt.xlabel(f"火点发生时间 - {crop_name}成熟日（天）", fontsize=12)
    plt.ylabel("火点数量", fontsize=12)
    plt.title(f"{crop_name}种植区火点与成熟期的时间差分布", fontsize=14)
    
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()


# --------------------------
# 3. 调用绘图函数（保持原有逻辑）
# --------------------------
# 绘制玉米时间差图
if 'maize_valid' in locals():
    plot_time_diff(maize_valid, "maize")
else:
    print("⚠️ 未检测到maize_valid数据，跳过玉米时间差图绘制")

# 绘制小麦时间差图
if 'wheat_valid' in locals():
    plot_time_diff(wheat_valid, "wheat")
else:
    print("⚠️ 未检测到wheat_valid数据，跳过小麦时间差图绘制")
```

```python
def plot_yearly_diff(valid_diff, crop_type="maize"):
    """分年份绘制时间差箱线图"""
    diff_col = f"days_after_{crop_type}"
    # 按年份分组
    yearly_data = [
        valid_diff[valid_diff["year"] == str(y)][diff_col] 
        for y in sorted(valid_diff["year"].unique().astype(int))
    ]
    
    plt.figure(figsize=(12, 6))
    plt.boxplot(yearly_data, labels=sorted(valid_diff["year"].unique()))
    plt.axhline(y=1, color="r", linestyle="--", alpha=0.5)
    plt.axhline(y=30, color="r", linestyle="--", alpha=0.5)
    plt.ylabel(f"火点时间差（天）", fontsize=12)
    plt.title(f"{crop_type}种植区火点与成熟期时间差的年际变化", fontsize=14)
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.show()

# 玉米分年份对比
if 'maize_valid' in locals():
    plot_yearly_diff(maize_valid, "maize")

# 小麦分年份对比
if 'wheat_valid' in locals():
    plot_yearly_diff(wheat_valid, "wheat")
```

```python
# --------------------------
# 第一步：导入所有需要的库
# --------------------------
import os
import pandas as pd
import geopandas as gpd
import rasterio
from rasterio.transform import rowcol
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# --------------------------
# 第二步：修正数据路径（关键！完全匹配你的实际路径）
# --------------------------
base_dir = "/Users/leo/Desktop/CNGF5020/MINIPRO/Mini Group Project I Data(1)"
fire_dir = os.path.join(base_dir, "Satellite Fire Data")  # 卫星火点数据目录（不变）
county_shp_path = os.path.join(base_dir, "CHN_County.shp")  # 黑龙江省边界文件（不变）
# 🔥 修正：玉米/小麦物候TIFF直接在"Cropland..."目录下，无需额外子文件夹
crop_pheno_dir = os.path.join(base_dir, "Cropland distribution and phenological data")  # 物候数据总目录

# --------------------------
# 第三步：生成 maize_valid 和 wheat_valid（变量来源，路径已修正）
# --------------------------
# 1. 读取并处理黑龙江省内的火点数据（不变）
fire_csv_paths = [os.path.join(fire_dir, f) for f in os.listdir(fire_dir) if f.endswith(".csv") and "modis" in f]
fire_csv_paths.sort()
fire_df_list = []
for csv_path in fire_csv_paths:
    df = pd.read_csv(csv_path, usecols=['latitude', 'longitude', 'acq_date', 'acq_time', 'frp'])
    # 筛选黑龙江省大致范围
    df = df[(df['latitude'] >= 43) & (df['latitude'] <= 53) & (df['longitude'] >= 121) & (df['longitude'] <= 135)]
    # 处理时间和年份
    df['acq_time_str'] = df['acq_time'].astype(str).str.zfill(4)
    df['datetime'] = pd.to_datetime(df['acq_date'] + " " + df['acq_time_str'].str[:2] + ":" + df['acq_time_str'].str[2:])
    df['year'] = df['datetime'].dt.year.astype(str)
    fire_df_list.append(df)
fire_all_df = pd.concat(fire_df_list, ignore_index=True)

# 转成GeoDataFrame并精确裁剪到黑龙江省
fire_gdf = gpd.GeoDataFrame(
    fire_all_df,
    geometry=gpd.points_from_xy(fire_all_df['longitude'], fire_all_df['latitude']),
    crs="EPSG:4326"
)
# 读取黑龙江省边界
hlj_boundary_gdf = gpd.read_file(county_shp_path)
province_field = "NAME_1"
hlj_boundary_gdf = hlj_boundary_gdf[
    (hlj_boundary_gdf[province_field] == "Heilongjiang") | 
    (hlj_boundary_gdf[province_field] == "HEILONGJIANG")
]
fire_hlj_gdf = gpd.clip(fire_gdf, hlj_boundary_gdf)  # 黑龙江省内火点

# 2. 火点与作物匹配函数（不变）
def match_fire_with_crop(fire_gdf, crop_tif_paths, crop_type="maize"):
    matched_dfs = []
    for tif_path in crop_tif_paths:
        # 从TIFF文件名提取年份（如"Heilongjiang_Maize_2010.tif" → 2010）
        file_basename = os.path.basename(tif_path)
        year = int(file_basename.split("_")[-1].split(".")[0])  # 按文件名格式提取年份
        year_str = str(year)
        # 筛选该年份的火点
        fire_year = fire_gdf[fire_gdf["year"] == year_str].copy()
        if len(fire_year) == 0:
            print(f"⚠️ {year}年无{['玉米', '小麦'][crop_type=='wheat']}火点数据，跳过")
            continue
        # 读取作物物候TIFF
        with rasterio.open(tif_path) as src:
            transform = src.transform
            nodata = src.nodata
            # 计算火点在栅格中的行列号
            rows, cols = rowcol(transform, fire_year.geometry.x, fire_year.geometry.y)
            # 提取成熟日DOY
            maturity_doy = []
            for r, c in zip(rows, cols):
                if 0 <= r < src.height and 0 <= c < src.width:
                    val = src.read(1)[r, c]
                    maturity_doy.append(val if val != nodata else None)
                else:
                    maturity_doy.append(None)
            # 添加成熟日字段并过滤无效值
            fire_year[f"{crop_type}_maturity_doy"] = maturity_doy
            fire_crop = fire_year.dropna(subset=[f"{crop_type}_maturity_doy"])
            if len(fire_crop) == 0:
                print(f"⚠️ {year}年无{['玉米', '小麦'][crop_type=='wheat']}种植区火点，跳过")
                continue
            # 计算时间差（火点DOY - 成熟日DOY）
            fire_crop["fire_doy"] = fire_crop["datetime"].dt.dayofyear
            fire_crop[f"days_after_{crop_type}"] = fire_crop["fire_doy"] - fire_crop[f"{crop_type}_maturity_doy"]
            matched_dfs.append(fire_crop)
    if not matched_dfs:
        return None
    return pd.concat(matched_dfs, ignore_index=True)

# 3. 时间差分析函数（不变）
def analyze_time_diff(matched_gdf, crop_type="maize"):
    diff_col = f"days_after_{crop_type}"
    # 筛选合理时间范围（成熟前30天到成熟后60天，排除极端值）
    valid_diff = matched_gdf[(matched_gdf[diff_col] >= -30) & (matched_gdf[diff_col] <= 60)]
    print(f"\n{['玉米', '小麦'][crop_type=='wheat']}时间差统计：有效火点{len(valid_diff)}条")
    return valid_diff

# 4. 🔥 修正：从物候总目录筛选玉米/小麦的TIFF文件（按文件名区分）
# 假设玉米TIFF文件名含"Maize"（如"Heilongjiang_Maize_2010.tif"），小麦含"Wheat"
maize_tif_paths = [
    os.path.join(crop_pheno_dir, f) 
    for f in os.listdir(crop_pheno_dir) 
    if f.endswith(".tif") and "Maize" in f  # 按"Maize"关键词筛选玉米TIFF
]
wheat_tif_paths = [
    os.path.join(crop_pheno_dir, f) 
    for f in os.listdir(crop_pheno_dir) 
    if f.endswith(".tif") and "Wheat" in f  # 按"Wheat"关键词筛选小麦TIFF
]

# 生成 maize_valid（玉米火点有效数据）
print("=== 处理玉米火点数据 ===")
fire_maize_gdf = match_fire_with_crop(fire_hlj_gdf, maize_tif_paths, "maize")
maize_valid = analyze_time_diff(fire_maize_gdf, "maize") if fire_maize_gdf is not None else None

# 生成 wheat_valid（小麦火点有效数据）
print("\n=== 处理小麦火点数据 ===")
fire_wheat_gdf = match_fire_with_crop(fire_hlj_gdf, wheat_tif_paths, "wheat")
wheat_valid = analyze_time_diff(fire_wheat_gdf, "wheat") if fire_wheat_gdf is not None else None

# --------------------------
# 第四步：绘制火点类型地图（直接调用有效变量）
# --------------------------
# 1. 火点类型分类函数（不变）
def classify_fire_types(matched_gdf, crop_type="maize"):
    classified_gdf = matched_gdf.copy()
    diff_col = f"days_after_{crop_type}"
    # 分类：疑似秸秆焚烧（成熟后1-30天）vs 非农业焚烧
    classified_gdf["fire_type"] = "非农业焚烧"
    crop_name_cn = "玉米" if crop_type == "maize" else "小麦"
    classified_gdf.loc[
        (classified_gdf[diff_col] >= 1) & (classified_gdf[diff_col] <= 30),
        "fire_type"
    ] = f"疑似{crop_name_cn}秸秆焚烧"
    return classified_gdf

# 2. 地图绘制函数（不变）
def plot_fire_type_map(classified_gdf, crop_type="maize"):
    # 配置中文字体（避免乱码）
    plt.rcParams["font.family"] = ["Arial Unicode MS", "Heiti TC", "SimHei"]
    plt.rcParams["axes.unicode_minus"] = False
    
    # 创建画布
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # 绘制黑龙江省边界（合并为省级，更清晰）
    hlj_province_gdf = hlj_boundary_gdf.dissolve()
    hlj_province_gdf.boundary.plot(
        ax=ax,
        color="black",
        linewidth=1.5,
        label="黑龙江省边界"
    )
    # 填充省份背景（浅灰色，突出火点）
    hlj_province_gdf.plot(ax=ax, color="#F5F5F5", alpha=0.5)
    
    # 火点颜色配置（秸秆焚烧：红色，非农业：蓝色）
    crop_name_cn = "玉米" if crop_type == "maize" else "小麦"
    color_map = {
        f"疑似{crop_name_cn}秸秆焚烧": "#E74C3C",  # 深红色
        "非农业焚烧": "#3498DB"                       # 蓝色
    }
    
    # 按类型绘制火点
    for fire_type, color in color_map.items():
        subset = classified_gdf[classified_gdf["fire_type"] == fire_type]
        subset.plot(
            ax=ax,
            color=color,
            markersize=6,
            alpha=0.7,  # 透明度避免重叠
            label=fire_type
        )
    
    # 图表美化（标题、标签、图例）
    ax.set_title(
        f"黑龙江省{crop_name_cn}种植区火点类型分布",
        fontsize=16,
        pad=20  # 标题与图的距离
    )
    ax.set_xlabel("经度", fontsize=12)
    ax.set_ylabel("纬度", fontsize=12)
    ax.grid(True, linestyle="--", alpha=0.4)  # 辅助网格线
    # 图例放在右侧，避免遮挡火点
    ax.legend(
        loc="upper right",
        fontsize=11,
        title="火点类型",
        title_fontsize=13,
        bbox_to_anchor=(1.35, 1)
    )
    
    # 调整布局并显示
    plt.tight_layout()
    plt.show()
    
    # 保存地图到项目目录（方便后续用）
    save_path = os.path.join(base_dir, f"{crop_name_cn}_fire_type_map.png")
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    print(f"✅ {crop_name_cn}火点地图已保存至：{save_path}")

# --------------------------
# 第五步：调用地图函数（变量已生成，无报错）
# --------------------------
# 绘制玉米火点地图
if maize_valid is not None and not maize_valid.empty:
    maize_classified = classify_fire_types(maize_valid, "maize")
    plot_fire_type_map(maize_classified, "maize")
else:
    print("⚠️ 玉米火点数据（maize_valid）为空或未生成，无法绘制地图")

# 绘制小麦火点地图
if wheat_valid is not None and not wheat_valid.empty:
    wheat_classified = classify_fire_types(wheat_valid, "wheat")
    plot_fire_type_map(wheat_classified, "wheat")
else:
    print("⚠️ 小麦火点数据（wheat_valid）为空或未生成，无法绘制地图")
```
